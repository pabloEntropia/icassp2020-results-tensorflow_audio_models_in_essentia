# Display and parse Jamendo annotations
This notebook contains manual annotations of the split-0 test set from the mtg-jamendo dataset.
Each track is annotated according to the taxonomies of the MTGDB collections

```python
import json
import os
import glob
import pandas as pd
from collections import Counter
```

```python
genre_rosamerica_groundtruth = 'groundtruth/groundtruth_rosamerica.json'
genre_tzanetakis_groundtruth = 'groundtruth/groundtruth_tzanetakis.json'
genre_dortmund_groundtruth   = 'groundtruth/groundtruth_dortmund.json'
genre_electronic_groundtruth = 'groundtruth/groundtruth_electronic.json'

genre_rosamerica = dict()
genre_dortmund = dict()
genre_gtzan = dict()
genre_electronic = dict()

os.chdir("jamendo_annotations/genre")

for file in glob.glob("*.json"):
    filename = file.split('/')[-1]
    task, _, page, _, idx = filename.split('-')
    idx = idx.replace('.mp3.json', '')
    data = json.load(open(file))
    
    if task == 'genre':
        genre_rosamerica[idx] = data[0]['genre_rosamerica']
        genre_dortmund[idx] = data[2]['genre_dortmund']
        genre_gtzan[idx] = data[1]['genre_tzanetakis']
        genre_electronic[idx] = data[3]['genre_electronic']

os.chdir("../../")       

json.dump({idx: [[val, 100]] for idx, val in genre_rosamerica.items()}, open(genre_rosamerica_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in genre_gtzan.items()}, open(genre_tzanetakis_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in genre_dortmund.items()}, open(genre_dortmund_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in genre_electronic.items()}, open(genre_electronic_groundtruth, 'w'))
```

```python
data = {'Genre Rosamerica': genre_rosamerica,
        'Genre Dortmund': genre_dortmund,
        'Genre Gtzan': genre_gtzan, 
        'Genre Electronic': genre_electronic}

for t, d in data.items():
    plt.bar(*zip(*dict(Counter(d.values())).items()))
    plt.xticks(rotation=45)
    plt.title(t)
    plt.show()
```

```python
mood_acoustic_groundtruth = 'groundtruth/groundtruth_mood_acoustic.json'
mood_electronic_groundtruth = 'groundtruth/groundtruth_mood_electronic.json'
mood_aggressive_groundtruth = 'groundtruth/groundtruth_mood_aggressive.json'
mood_relaxed_groundtruth = 'groundtruth/groundtruth_mood_relaxed.json'
mood_happy_groundtruth = 'groundtruth/groundtruth_mood_happy.json'
mood_sad_groundtruth = 'groundtruth/groundtruth_mood_sad.json' 
mood_party_groundtruth = 'groundtruth/groundtruth_mood_party.json'


mood_acoustic = dict() 
mood_electronic = dict() 
mood_aggressive = dict() 
mood_relaxed = dict() 
mood_happy = dict() 
mood_sad = dict() 
mood_party = dict() 

os.chdir("jamendo_annotations/mood")
for file in glob.glob("*.json"):
    filename = file.split('/')[-1]
    task, _, page, _, idx = filename.split('-')
    idx = idx.replace('.mp3.json', '')
    data = json.load(open(file))

    if task == 'mood':
        mood_acoustic[idx] = data[0]['mood_acoustic']
        mood_electronic[idx] = data[1]['mood_electronic']
        mood_aggressive[idx] = data[2]['mood_aggressive']
        mood_relaxed[idx] = data[3]['mood_relaxed']
        mood_happy[idx] = data[4]['mood_happy']
        mood_sad[idx] = data[5]['mood_sad']
        mood_party[idx] = data[6]['mood_party']

os.chdir("../../")

json.dump({idx: [[val, 100]] for idx, val in mood_acoustic.items()}, open(mood_acoustic_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in mood_electronic.items()}, open(mood_electronic_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in mood_aggressive.items()}, open(mood_aggressive_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in mood_relaxed.items()}, open(mood_relaxed_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in mood_happy.items()}, open(mood_happy_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in mood_sad.items()}, open(mood_sad_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in mood_party.items()}, open(mood_party_groundtruth, 'w'))
```

```python
data = {'Mood Accoustic': mood_acoustic,
        'Mood Electronid': mood_electronic,
        'Mood Aggresive': mood_aggressive,
        'Mood Relaxed': mood_relaxed, 
        'Mood Happy': mood_happy,
        'Mood Sad': mood_sad,
        'Mood Party': mood_party}

for t, d in data.items():
    plt.bar(*zip(*dict(Counter(d.values())).items()))
    plt.xticks(rotation=45)
    plt.title(t)
    plt.show()
```

```python
timbre_groundtruth = 'groundtruth/groundtruth_misc_timbre.json'
tonal_atonal_groundtruth = 'groundtruth/groundtruth_misc_tonal_atonal.json'
danceability_groundtruth = 'groundtruth/groundtruth_misc_danceability.json'
voice_instrumental_groundtruth = 'groundtruth/groundtruth_misc_voice_instrumental.json'
gender_groundtruth = 'groundtruth/groundtruth_misc_gender.json'

misc_timbre = dict()
misc_tonal_atonal = dict()
misc_danceability = dict()
misc_voice_instrumental = dict()
misc_gender = dict()

os.chdir("jamendo_annotations/misc")

for file in glob.glob("*.json"):
    filename = file.split('/')[-1]
    task, _, page, _, idx = filename.split('-')
    idx = idx.replace('.mp3.json', '')
    data = json.load(open(file))
    
    if task == 'miscellaneous':
        misc_timbre[idx] = data[0]['timbre']
        misc_tonal_atonal[idx] = data[1]['tonal_atonal']
        misc_danceability[idx] = data[2]['danceability']
        misc_voice_instrumental[idx] = data[3]['voice_instrumental']
        misc_gender[idx] = data[4]['gender']

os.chdir("../../")
        
json.dump({idx: [[val, 100]] for idx, val in misc_timbre.items()}, open(timbre_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in misc_tonal_atonal.items()}, open(tonal_atonal_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in misc_danceability.items()}, open(danceability_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in misc_voice_instrumental.items()}, open(voice_instrumental_groundtruth, 'w'))
json.dump({idx: [[val, 100]] for idx, val in misc_gender.items()}, open(gender_groundtruth, 'w'))
```

```python
data = {'timbre': Counter(misc_timbre.values()),
        'atonal/tonal': Counter(misc_tonal_atonal.values()),
        'danceability': Counter(misc_danceability.values()),
        'voice/instrumental': Counter(misc_voice_instrumental.values()), 
        'gender': Counter(misc_gender.values())}

# Rename keys
data['timbre']['dark'] = data['timbre'].pop('0')
data['timbre']['bright'] = data['timbre'].pop('1')


data['atonal/tonal']['atonal'] = data['atonal/tonal'].pop('0')
data['atonal/tonal']['tonal'] = data['atonal/tonal'].pop('1')

data['danceability']['not danceable'] = data['danceability'].pop('0')
data['danceability']['danceable'] = data['danceability'].pop('1')

data['voice/instrumental']['instrumental'] = data['voice/instrumental'].pop('0')
data['voice/instrumental']['voice'] = data['voice/instrumental'].pop('1')

data['gender']['instrumental'] = data['gender'].pop('2')
data['gender']['female'] = data['gender'].pop('1')
data['gender']['male'] = data['gender'].pop('0')

for t, d in data.items():
    plt.bar(*zip(*dict(d).items()))
    plt.xticks(rotation=45)
    plt.title(t)
    plt.show()
```
